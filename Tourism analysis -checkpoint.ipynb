#### Importing library and reading file from csv while calculating the descriptive, counting the null value, listing the info of the data and replacing null value with zero and median

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mtc= pd.read_csv('C:\\Users\\MA1606\\OneDrive - Canterbury Christ Church University\\Documents\\python\\monthly_passengers.csv')
```

```python
mtc
```

```python
mtc['Month'] = pd.to_datetime(mtc['Month'], format='%m').dt.strftime('%b')
```

```python
mtc
```

```python
mtc.isnull().sum()
```

```python
mtc.fillna(0)
```

```python
# Fill missing values with median
for col in ['Total', 'Domestic', 'International']:
    mtc[col].fillna(mtc[col].median(), inplace=True)
```

```python
mtc
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=mtc, x= 'Month', y= 'Total_OS', hue= 'Year', marker='o')
plt.title('Monthly passenger Trend')
plt.xlabel("Month")
plt.ylabel("Total passengers")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=mtc, x= 'Month', y= 'International', hue= 'Year', marker='o')
plt.title('International Monthly passenger Trend')
plt.xlabel("Month")
plt.ylabel("International passengers number")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=mtc, x= 'Month', y= 'Domestic', hue= 'Year', marker='o')
plt.title('Domestic Monthly passenger Trend')
plt.xlabel("Month")
plt.ylabel("Domestic passengers number")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# Top 10 Countries by Total Passengers

```python
top_countries = mtc.groupby('ISO3')['Total_OS'].sum().nlargest(10).reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=top_countries, x='ISO3', y='Total_OS', palette='viridis')
plt.title("Top 10 Countries by Total Passengers")
plt.xlabel("Country")
plt.ylabel("Total Passengers")
plt.show()
```

```python
top_countries = mtc.groupby('ISO3')['International'].sum().nlargest(10).reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=top_countries, x='ISO3', y='International', palette='viridis')
plt.title("Top 10 Countries by International Passengers")
plt.xlabel("Country")
plt.ylabel("Total International Passengers")
plt.show()
```

```python
yearly_passengers = mtc.groupby(['ISO3', 'Year'])['Total_OS'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=yearly_passengers, x='Year', y='Total_OS', palette='viridis')
plt.title("Top 10 Countries by Total Passengers")
plt.xlabel("Year")
plt.ylabel("Total Passengers")
plt.show()
```

# ---------------- 3. DATA AGGREGATION & GROUPING ----------------

##### Top 5 Countries in Passenger Growth (Year-over-Year Comparison)

```python
yearly_passengers = mtc.groupby(['ISO3', 'Year'])['Total_OS'].sum().reset_index()
growth = yearly_passengers.pivot(index='ISO3', columns='Year', values='Total_OS')
growth['Growth'] = growth.iloc[:, -1] - growth.iloc[:, -2]  # Last year - Previous year
top_growth = growth.nlargest(5, 'Growth').reset_index()
print("Top 5 Countries by Passenger Growth:")
print(top_growth[['ISO3', 'Growth']])
```

```python
yearly_passengers = mtc.groupby(['ISO3', 'Year'])['International'].sum().reset_index()
growth = yearly_passengers.pivot(index='ISO3', columns='Year', values='International')
growth['Growth'] = growth.iloc[:, -1] - growth.iloc[:, -2]  # Last year - Previous year
top_growth = growth.nlargest(5, 'Growth').reset_index()
print("Top 5 Countries by Passenger Growth:")
print(top_growth[['ISO3', 'Growth']])
```
