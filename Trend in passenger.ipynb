##### Now we want to calculate Monthly Trend yearly for total passenger arrival. 

```python
plt.figure(figsize=(14,6))
sns.lineplot(data=mtc, x='Month', y= 'Total', hue='Year', marker='o')
plt.title("Total Monthly passenger Trend")
plt.xlabel("Month")
plt.ylabel("Total passenger")
plt.legend(rotation=45, title='Year')
plt.show()
```
